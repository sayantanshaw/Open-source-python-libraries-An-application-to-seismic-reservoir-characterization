```python
class Operator(object):
    

    def __init__(self, wavelet):
         
        self.wavelet= wavelet
         
        
    def forward(self, r):
         
        return np.convolve(r, self.wavelet, mode ='same') 
                           
    def adjoint(self, r):
         
        return np.correlate(r, self.wavelet, mode ='same') 
```
